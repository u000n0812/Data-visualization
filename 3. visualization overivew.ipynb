

### 분포 시각화

- histogram과 density plot은 분포 데이터를 가장 직관적으로 보여준다. 다만 파라미터를 정해야 해서 데이터가 잘못 전달될 위험도 있다.

- 이럴 때 cumulative density와 quantile-quantle plot을 활용해 데이터를 정확하게 표현해줄 수 있다. 다만 이들은 해석이 까다롭다.

- 이 외에도 box-plot, violin plot, strip chart, sina chart 등은 여러 분포 값을 한번에 나타내어 비교하거나 변화 추이를 한눈에 볼 수 잇다.

- stacked histogram과 overlapped density plot 등은 데이터 수가 적을 때 심층 비교 분석을 하기 좋다.

- Ridgeline plot은 violin plot 대신 쓰기 좋고, 데이터가 아주 많은 경우 또는 시간의 흐름에 따라 분포가 달라지는 경우에 특히 유용하다.

### X-Y 관계로 나타내는 시각화

- scatter plot은 하나의 정량적 변수를 다른 변수와 비교할 때 많이 활용된다.

- bubble chart는 정량적 변수가 3개일 때 변수 하나를 크기의 기준으로 삼는 방법이다.

- 만약 x축과 y축의 단위가 동일하거나 쌍을 이루는 관계면 y=x 축을 그려주는게 해석에 도움이 된다.

- 비슷하게 쌍을 이루는 데이터를 표현할 때는 slopegraph도 도움이 된다.


```python
Image('D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/directory_of_visualizations_files/figure-html/basic-scatter-1.png')
```

- 데이터 포인트가 너무 많은 데이터셋은 제대로 알아볼 수가 없기 때문에 contour plot, 2D Bins, Hex Bins 등을 활용한다. 다만 시각화하려는 데이터가 2가지 이상이면 Correlogram을 활용하는 것도 좋다.

```python
Image('D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/directory_of_visualizations_files/figure-html/xy-binning-1.png')
```

### 지리공간 데이터 시각화

- 지리공간 데이터는 기본적으로 지도 위에 위치한다. 지도는 지구본 좌표를 평면에 옮겨서 2차원 공간에서 거리를 나타낸 것이다. 지도에 색을 칠해서 짖역별 데이터 값을 보여줄 수 있는데 이런 지도를 choropleth라고 부른다. 그리고 데이터 값에 따라 지역 형태를 왜곡하거나, 지역 모양을 단순화하는 방식을 cartogram이라고 부른다.

### 불확실성의 시각화

- 오차 막대(error bar)는 추정하거나 측정한 데이터의 예상 값 범위를 나타낸다. 예상 범위는 가로나 세로 방향으로 확장하며, 점이나 막대 등 다양한 형태로 표현될 수 있다. graded error bars는 여러 예상 범위를 동시에 나타내지만, 범위마다 신뢰도가 달라 헷갈릴 수 있다.

```python
Image('D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/directory_of_visualizations_files/figure-html/errorbars-1.png')
```

- 더 상세하게 실제 신뢰도나 posterior distribution을 시각화할 수도 있다. 이때 confidence strip은 불확실성은 쉽게 볼 수 있지만 정확한 수치를 읽기 어렵다.

- 눈 모양의 eyes graph는 error bar에 분포(violin 등)를 결합한 것이다.

- quantile dot plot은 불확실성 분포도 대신 활용할 수 있으며, 별개 단위를 사용하기에 아주 정확하지는 않지만 동일한 데이터를 연속된 형태로 보여주는 violin이나 ridgeline보다는 이해하기가 쉽다.

```python
Image('D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/directory_of_visualizations_files/figure-html/confidence-dists-1.png')
```
