# principle of proportional ink

- 그래픽 요소를 사용할 때는 크기, 형태, 위치 등에 따라 문제가 생긴다. 모순이 생기지 않도록 주의해야 하는데, 이를 principle of proportional ink라고 부른다.

- 작은 변화를 나타내려고 할 때는 막대나 영역 색칠이 유용하지 않다고 생각할 수 있다. 아래와 같이 y축의 범위에 따라 변하는 정도가 극적으로 바뀔 수 있기 때문이다.

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/proportional_ink_files/figure-html/fb-stock-drop-bad-1.PNG")
```

- 하지만 어떤 차이를 보여주려고 하는지에 따라 조건별로 선택하면 더욱 차이를 잘 보여줄 수 있다. 예를 들어 정량값이 아니라 변동을 시각화하는 것으로 바꾸면 아래와 같이 더 확실한 차이를 보여줄 수 있거나, 상승세가 아니라 하락세를 표기하는 등이다.

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/proportional_ink_files/figure-html/hawaii-income-change-1.PNG")
```

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/proportional_ink_files/figure-html/fb-stock-drop-inverse-1.PNG")
```

- 일반적으로 데이터 값과 도형이 차지하는 면적은 비례하지만, log 변환을 했으면 그렇지 않다. log 값에 비례하기 때문이다. 

- 앞서 log 스케일은 비율을 시각화하기에 알맞다고 했으나, 실제로는 도표에 나타낼 숫자의 자릿수가 다양한 경우에 많이 쓰인다. 예를 들어 오세아니아의 GDP는 10억 달러 미만부터 3000억 달러 이상까지 다양한데, 선형 스케일로 시각화하는 것이 효과적이지 않기 때문에 로그 스케일을 사용한다.

- 그러나 막대가 임의의 값부터 출발하면 여전히 문제가 생긴다. 게다가 로그 변환했기에 무작정 0부터 그릴 수도 없다. 그래서 스케일을 따라 국가별로 해당 위치에 점을 찍어서 GDP를 찍는 방법으로 대체하는 경우가 많다. 아래와 같이 국가명을 y축이 아니라 점 바로 옆에 표기했기 때문에 막대 길이 문제에서 벗어날 수 있다.

# Overplotting

- 데이터가 너무 많으면 scatter plot에서 서로 위치가 겹치는 문제가 생긴다. 이를 overplotting이라고 한다.

- 이런 경우 첫 번째 해결책은 투명도를 조절하는 것이다. 점마다 반투명 값을 넣으면 겹친 부분의 음영 차이로 인해 점이 얼마나 겹쳐 있는지가 파악된다.

- 그걸로만 해서 부족하면 겹치는 점들을 조금씩만 이동시키는 jittering도 유용하다. 단, 이 경우 데이터 값이 달라진다는 점에 주의하자. 너무 많이 옮기면 오해의 소지가 생긴다.

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/overlapping_points_files/figure-html/mpg-cty-displ-jitter-1.PNG")
```

- 그러나 투명도와 jittering도 압도적인 데이터들에는 쓰기 불편하다. 이럴 땐 점을 하나하나씩 찍는 대신 2차원 히스토그램으로 표현하는 방법이 있다. 평면을 작은 직사각형으로 나누고, 그 영역 하나에 관측 값이 몇 개씩 들어가는지에 따라 색을 칠하는 방법이다.

- 직사각형보다 육각형은 점들로부터 중심까지 거리가 더 가까워서 데이터를 조금 더 정확하게 표현할 수 있다.

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/overlapping_points_files/figure-html/nycflights-hex-bins-1.PNG")
```

- 또는 도표 전체의 점 밀도를 추정해 등고선을 그릴 수도 있다. 점 밀도 비슷한 영역을 표시하는 등고선 위에 투명도를 조절한 점을 찍으면 분포 상태가 더 선명하게 드러난다.

# 효과적인 색 사용

- 너무 많은 범주에 일일이 색을 부여하면 도표가 너무 과해진다. 도표와 범례를 왔다갔다 하는 것은 여간 어려운 일이 아니다. 이런 방식은 색을 사용하는 목적에 어긋난다.

- 일반적으로 3~5개의 색상 스케일이 적당하다. 그 이상이면 너무 많아서 효과가 떨어진다. 구분해야할 범주가 8개 이상이면 색을 입히는 대신, 색은 다른 범주를 만들어 입히고 대표적인 데이터 옆에 레이블을 다는 편이 낫다.

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/pitfalls_of_color_use_files/figure-html/popgrowth-vs-popsize-bw-1.PNG")
```

- 적당한 수의 색을 사용한다 하더라도, 목적 없이 사용하는 것은 좋지 않다. 색을 의미 없이 사용하거나, 너무 강하고 튀게 사용한 현란한 도표는 오래 보기가 힘들다.

- 색상은 데이터 값의 크고 작음을 명확하게 알려줘야 하며, 색상 차이는 데이터 값 차이에 부합하여 시각화되어야 한다. 

- 도표에 쓸 색을 고를 때는 언제나 색약/색맹 환자도 고려하자. 색각 이상을 겪는 사람은 색을 전혀 구별하지 못할 수도 있다. 특히 빨강과 녹색(적록 색맹), 파랑과 녹색(청황 색맹)같은 경우가 많다.

- 앞전에 설명한 것처럼 시각화에 사용되는 색상 스케일은 순차적, 발산형, 정성적이라는 3가지 유형이 있다. 이중 색각 관련 문제가 가장 적은 방식은 순차적 스케일이다. 어두운 색부터 밝은 색 순으로 그라데이션이 나타나기 때문이다.

- 발산형은 일이 복잡해진다. 색각 이상이 없는 사람에게는 빨강과 녹색이 가장 강하게 대비되지만, 적색각, 녹색각 이상자에게는 구별되지 않는 등이다.

- 색각 이상자에게 안전하면서 대비되는 두 색을 차장내기란 불가능에 가깝다. 하지만 치명적인 문제는 아니다. 색각 이상자가 보기에도 지장이 없으면서 필요한 속성이 강조되도록 색을 조정하면 되기 때문이다. 아래에서 분홍색-연두색 스케일은 비이상자에게는 빨강-녹색으로 보이면서, 색각 이상자도 색을 구별할 수 있다.

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/color_basics_files/figure-html/diverging-scales-1.png")
```

- 가장 까다로운건 정석적 스케일이다. 각기 다른 여러 색이 필요하면서, 모든 종류의 색각 이상자가 구별할 수 있는 색이어야 하기 때문이다. 그래서 색각 이상 관련 문제를 돌파하기 위해 특별히 개발한 정성적 스케일은 아래와 같다. 이 스케일은 각기 다른 8개 색상으로 이루어져 있어서 대부분 상황에서 구별해가며 사용할 수 있다.

- 다만 색각 이상자에게 적절한 색상 스케일이 꽤 제작되어 있지만, 만능 해결사로 여기지는 말아야 한다. 어떻게 해도 알아볼 수 없는 도표가 나올 수도 있기 때문이다. 그렇기에 색을 넣는 그래픽 요소의 크기에 신경쓰고, 넓은 면적에 색을 입혀 구별이 쉽게 만드는 것이 중요하다. 색각 이상자에게는 그 차이가 더 크게 느껴진다.

- 온라인 서비스나 컴퓨터 프로그램을 써서 원하는 도표에 대해 색각 이상 시뮬레이션을 돌려보자.

# 불필요한 기호화

- 색을 넣는다고 무조건 정보를 효과적으로 전달할 수 있는건 아닌 것처럼, 굳이 필요 없는 기호를 추가하는 디자인 방식을 불필요한 기호화(redundant coding)이라고 부른다.

- 색약의 문제를 조금 이어가자면, 적녹 색약이 있는 사람들은 의외로 녹색과 파란색은 뚜렷하게 구별한다. 반면 파란색과 노란색에 대한 색약이 있는 사람에게는 파란색과 녹색이 비슷하게 보인다. 이런 문제를 해결하는 2가지 방법이 있다.

- 첫 번째로, 색상 순서를 바꿔서 파란색과 녹색 점이 서로 붙어 있지 않게 만드는 방법이다.

- 두 번째로, 3가지 도형을 활용해 점들의 모양을 다르게 표현하는 것이다.

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/redundant_coding_files/figure-html/iris-scatter-three-shapes-1.png")
```

- scatter plot에서는 점 모양을 바꿀 수 있지만, line plot에는 그것이 어렵다. 점선이나 쇄선으로 표현해도 잘 구별이 안된다. 이런 경우 범례의 순서와 그래프의 높이 순서를 일치시키면 그나마 보기가 쉬워진다. 즉, 데이터에 시각적인 순서가 존재하면 범례도 순서를 맞춰주는 방법이다.

- 이는 늘 지켜야 하는 원칙이긴 하지만, 색각 이상 시뮬레이션을 돌려보면 그 중요성이 더 확실하게 드러난다.
