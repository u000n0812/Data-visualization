### 데이터 값 자체를 색으로 표현하기

- 데이터 값 자체를 색상으로 나타내려면 주로 색에 순서를 부여해 값의 크기나 거리를 명확하게 보여준다. 색상 스케일을 이루는 색들이 전체적으로 균일하게 차이를 보여야 두 값 사이의 거리가 잘 표현된다.

- 순차적인 색상을 부여할 때는 하나의 색조만 사용하는 방식(예를 들어 파랑 계열), 여러 색조를 사용하는 방식(빨강과 노랑 계열)으로 나눌 수 있다. 여러 색조를 넣는 스케일은 보통 자연물에서 볼 수 있는 그라데이션을 따를 때가 많다. 색의 순서가 뒤집히면 보기에 부자연스럽고 순차적 개념도 명확하게 드러나지 않는다.

- 이 방법은 지역별 데이터 값을 나타낼 때 특히 유용하다. 지역의 지도에 색을 넣은 그림을 단계구분도(choropleth)라고 부르는데, 아래 Figure 2와 같다.

- 경우에 따라 중간점을 두고 방향을 지정해야 할 수도 잇다. 예를 들어, 0을 기준으로 양수/음수는 물론이고 0으로부터의 편차가 얼마인지도 파악해야 하는 경우이다. 이럴 때 쓰기 좋은 스케일로 발산형 스케일(diverging color scale)이 있다. 이는 순차적 스케일 2개를 공통의 중간점에서 연결한 스케일로 볼 수 있다. 이때 중간점은 보통 연한 색으로 나타낸다.

- 발산형 스케일은 색의 균형이 잘 맞아야 한다. 중간점을 출발해 양쪽으로 전개되는 두 색이 균일한 차이를 보이면서 점점 짙어져야 하는 것이다.

- 발산형 색상 스케일 예시는 아래 Figrue 3-1과 같고, 이를 활용한 지도는 Figure 3-2에서 확인할 수 있다.
