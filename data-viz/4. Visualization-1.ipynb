# 분포 시각화

- 모두 같은 구간을 설정해 데이터를 나타내는 histogram이 가장 대표적인 분포 시각화 그래프이다. 다만 구간의 폭을 정확하게 설정해야 시각적으로도 정확한 결과물이 나오기 때문에 optimal한 폭을 찾아내야 한다.

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/visualizing_distributions_I_files/figure-html/titanic-ages-hist-grid-1.png")
```

- 요즘은 histogram 대신 density plot을 그리는 추세다. 연속적인 곡선으로 확률 분포를 나타내는데, kernel density estimation 방법을 가장 많이 활용해 plot을 그린다. 이 방법은 데이터 포인트 지점마다 일정한 대역폭이라는 파라미터로 커널(곡선)을 그린 다음, 전부 더해서 최종 밀도 추정 값을 구하는 방법이다. Gaussian kernel을 자주 활용하지만, 다른 선택지도 많다.

- histogram과 마찬가지로 대역폭을 어떻게 설정하느냐에 따라 plot의 모양이 결정된다. density plot은 아래 면적이 1이 되도록 그리는데, x축과 y축의 단위에 따라 헷갈릴 수도 있으니 조심하자.

- 그리고 KDE에서 일부 영역, 특히 꼬리 부분에 데이터 값이 존재하지 않는 것처럼 보이기 쉽다. 그래서 주의를 기울이지 않으면 앞뒤가 안맞는 plot이 그려질 수 있다.

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/visualizing_distributions_I_files/figure-html/titanic-ages-dens-grid-1.png")
```

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/visualizing_distributions_I_files/figure-html/titanic-ages-dens-negative-1.png")
```

- 그래서 그 대신, point cloud 형태로 데이터 포인트를 전부 표시하는 방법도 있다. 다만 이렇게 하면 데이터가 너무 많거나 분포 속성을 강조하는 집계값을 표시하기 위해 ECDF(Empirical Cumulative Distribution Function)와 Q-Q plot(Quantile-Quantile plot)을 고안했다. 이 둘은 히스토그램이나 density plot처럼 직관적이지는 않으나, 파라미터를 설정하지 않아도 되기에 꽤나 선호도가 높다.

- 여러 개의 분포 상태를 다룰 때는 항상 response와 group variable을 생각해야 한다. response는 분포 상태를 보여주려고 하는 중요 변수이고, grouping variable은 response의 분포를 뚜렷이 구분하기 위한 변수이다.
- 예를 들어, 기온이 response, month는 grouping variable이 된다.
- 이런 경우, boxplot, violin plot, ridgeline plot 등이 자주 활용된다.
