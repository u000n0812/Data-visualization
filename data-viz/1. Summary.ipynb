# Chapter 2. 위치 스케일 : 좌표, 축

- 위치 스케일은 그래프안에서 데이터 값의 위치를 결정하는 스케일을 말한다. 일반적으로 x축, y축으로 만들지만, 서로 예각이 되게 만드는 등 다양하게 만들 수 있다. 이런 위치 스케일을 모든 집합과 스케일들의 배열 관계를 **좌표계(coordinate system)**라고 부른다.

- 그 중 x값과 y값을 조합해 위치 값을 명시하는 **데카르트 좌표계(Cartesian coordinate system)**, 직교 좌표계가 자주 활용된다. 데카르트 좌표계에서 격자선은 각 데이터 단위가 어느 정도로 나누면 적절할지에 따라 나눌 수 있다. 만약 x축, y축의 단위가 다른 경우 한 축의 데이터 값을 줄이거나 늘리면 반드시 다른 축의 값도 그에 비례하게 줄이거나 늘려야 데이터 시각화의 타당성을 유지할 수 있다.

- 이와 비슷하게, 그래프의 가로 세로 비율을 조절해서 한 축의 값이 너무 급격하거나 너무 완만하게 느껴지지 않도록 잘 조절해야 한다.

```python
Image("D:/뇌/2023/독서목록/데이터 시각화 교과서/code/docs/coordinate_systems_axes_files/figure-html/temperature-normals-Houston-1.png")
```

- x축과 y축의 단위가 같다면, 두 축의 격자선 가격을 통일하는 편이 좋다. 그래서 데이터 값이 같으면 그 값을 나타내는 길이 값도 같게 표현하는 것이 낫다. 이렇게 하면 혹시 데이터 단위가 달라지더라도, 축만 수정하면 그래프 형태는 완전히 똑같이 다시 만들 수 있다.

- 격자선을 그을 때 스케일 변환을 하기도 하는데, linear한 방식과 non-linear한 방식으로 나눌 수 있다. non-linear한 방식 중 가장 유명한 방법은 logarithmic scale로, 로그 변환된 데이터로 스케일을 표현하는 것이다. 이때, 축에 표시되는 눈금에 레이블 표시와 어떻게 변환했는지만 잘 바꿔주면 된다.

- 로그 스케일에서의 곱셈은 선형 스케일의 덧셈과 모양이 같기 때문에, 곱셈과 나눗셈으로 얻은 값을 로그 스케일로 표현하면 자연스럽게 표시할 수 잇다. 특히 비율 값에 로그 스케일을 자주 활용한다.

- 텍사스의 카운티별 인구를 나타낸 Figure 4-1과 Figrue 4-2를 비교해보자. 같은 데이터를 선형 스케일로만 옮겨 놓은 Figure 4-1은 인구 차이를 정확히 파악하기가 힘들다. 하지만 각 카운티별 인구를 전체 인구의 중간 값으로 나눈 Figure 4-2는 중간 값을 기준으로 대칭적으로 분포하며, 중간값보다 몇 배나 더 많은 인구가 어디에 살고 있는지를 한눈에 파악할 수 있다.

- 이처럼 logarithtic transformation은 데이터셋에 포함된 값들이 심하게 들쭉날쭉할 때 주로 사용한다.

- 하지만 0이라는 값이 포함되어야 하면 어떡할까? 그러면 로그를 취하면 음의 무한대가 되므로 로그 변환 대신 제곱근 스케일을 사용할 수 있다.

- 제곱근 스케일은 2가지 단점이 있다. (1) 선형 스케일에서 더하거나 빼고, 로그 스케일에서 곱하거나 나누는 것이 한 단위에 대응되는 것과 달리, 제곱근 스케일은 대응되는 것이 없다. (2) 축의 어느 위치에 눈금을 표시하는게 좋을지 애매하다.

- 그럼에도 제곱근 스케일은 대안책으로 자주 활용되며, 땅의 면적과 같이 애초부터 제곱 형태로 계산되는 데이터들의 경우 훨씬 효과적이다.
