# 이미지 형식

파일 형식마다 장단점이 있는데, 형식만 잘 정해도 도표 제작 과정에서 겪는 골치 아픈 문제들을 줄일 수 있다.  

주로 인쇄용 고화질 파일을 비롯한 대부분의 경우에는 PDF, 온라인 문서나 비트맵 그래픽이 필요한 경우는 PNG, PNG 파일이 너무 큰 경우에는 JPEG를 최종 수단으로 사용한다.

그래픽 형식을 구별하는 가장 중요한 차이점은 그 유형이 비트맵(bitmap)인지 벡터(vector)인지에 따라 달라진다.  

비트맵은 화소 또는 픽셀이라 불리는 점을 배열해 이미지를 완성하는 반면, 벡터 그래픽은 이미지를 구성하는 그래픽 요소들의 기하학적 배열 상태를 저장한다. 즉, 우층 상단에는 검은 색 선이 있고 ... 이런걸 저장한다.

vector graphic은 아무리 확대해도 디테일이나 선명도가 떨어지지 않아서 *'자유 해상도(resolution-independent)'*라고도 한다.  

다만 실제 작업 시 vector graphic은 파일을 열 때마다 새로 그리기 때문에 같은 그래픽도 다른 프로그램이나 컴퓨터에서 열면 다르게 보일 수 있으며(특히 폰트), 파일 크기가 너무 커지고 렌더링 속도가 느려질 수 있다는 단점이 있다.  

이와 달리 비트맵 이미지는 언제 어디에서 열어도 똑같은 모습으로 보인다.

비트맵 파일 형식은 용량을 적당 선에서 유지하기 위해 압축 방식을 채택한다. 주로 무손실과 손실로 나뉘는데, **무손실 압축**은 압축된 이미지와 원본의 화소가 일대일로 대응되는 반면, **손실 압축**은 용량을 줄이는 대신 이미지 화질을 어느 정도 저하시킨다.  

- **무손실 압축**  
검은 색은 (0, 0, 0)으로 저장된다. 그러면 검은색은 0이 3개 있다는 표현으로 바꿔 저장하면 용량을 줄일 수 있다. 숫자 2개만 저장해도 되기 때문이다. 이외에도 오랜 세월에 걸쳐 기발한 압축 방법들이 개발되었고, 요즘은 PNG 등의 무손실 이미지 형식을 효율성 높은 비트맵 파일로 저장할 수 있다. 다만 무손실 압축 알고리즘은 이미지에서 하나의 색이 넓은 면적을 차지할 때 가장 효과적이므로, 이 부분을 고려해야 한다.  
  
- **손실 압축**  
실제 사진은 명도나 색이 연달아 있는 경우가 거의 없고 무늬가 나타나므로 이런 경우에는 손실 압축을 활용한다. 손실 압축의 핵심은 이미지 품질을 눈에 띄게 저하시키지는 않으면서 눈으로 식별하기 힘든 미묘한 디테일을 버리는 것이다. 색상 값이 조금씩 다른 화소 1000개가 있을 때, 200개만 사용하고 인접한 화소를 5개씩 묶어 동일한 색을 부여하더라도 사람 눈에는 거의 똑같아 보이는 그라데이션을 만들 수 있다.  

가장 널리 활용되는 손실 이미지 형식은 JPEG이고, 디지털 카메라로 찍은 사진이 출력되는 기본 값도 보통 JPEG이다. JPEG 압축은 이미지 품질을 아주 조금만 저하시키면서 용량을 크게 줄일 수 있어 사진 저장에 특히 효과적이다. 그러나 선 그림이나 글자처럼 가장자리가 매끈한 이미지는 JPEG 압축이 적절하지 않다. 이 경우에 이미지를 JPEG로 압축하면 품질이 눈에 띄게 저하된다.  

JPEG로 변환된 이미지가 맨눈으로는 괜찮아 보여도 인쇄했을 때 선명도가 급격히 떨어질 수 있다. 따라서 되도록이면 피하는 것이 좋으며, 특히 시각화나 캡처 스크린샷처럼 선 그림이나 글자가 들어간 이미지는 PNG나 TIFF로 저장하는 편이 낫다. JPEG는 사진 이미지 저장에만 주로 사용된다.

대체로 이미지 형식 변환에는 제약이 없지만, 중요한 정보가 손실될 수 있고, 손실된 정보를 되돌릴 수 없기에 주의해야 한다. 예를 들어 vector graphic을 bitmap 형식으로 저장하면(PDF -> JPEG) vector graphic의 핵심인 자유 해상도가 사라진다. 반면 JPEG를 다시 PDF로 저장한다고 해서 다시 되돌아가지 않는다. 그 이미지는 여전히 PDF 안에 저장된 bitmap 이미지이다.  

마찬가지로 JPEG는 PNG로 변환해도 JPEG 과정에서 생긴 잡티가 사라지지는 않는다.  

따라서 경험에 따르면 항상 해상도, 정확도, 유연성을 극대화할 수 있는 형식으로 원본 이미지를 저장하는 것이 좋다. 시각화 결과물은 일단 PDF로 저장했다가 필요할 때 PNG나 JPEG로 변환하거나, 처음무터 고해상도 PNG 파일로 저장하는 편이 좋다. 디지털 사진처럼 비트맵으로만 저장할 수 있는 이미지인 경우는 무손실 아북 방식으로 저장하며, 손실 압축 방식을 선택해야 한다면 압축률을 최대한 낮춰서 저장한다. 또한 ㄴ이미지는 최대한 높은 해상도로 저장해뒀다가 필요할 때 압축해서 쓴다.

# 시각화 소프트웨어 선택

각자 자신에게 잘 맞는 프로그램을 선택하자. 큰 수고 없이 내가 원하는 도표를 만들 수 있으면 그걸로 땡이다.  

다만 작업 방식에 대한 상대적인 장점을 평가하는데 적용할 수 있는 일반적인 원칙은 존재한다.

실험에서 여러 연구진이 같은 연구를 진행할 때 핵심 결과가 달라지지 않으면 **재현 가능성(reproducibility)** 이 있다고 하고, 동일한 사람이 동일한 장비로 동일한 절차를 반복했을 때 동일한 결과를 얻을 수 있으면 **반복 가능성(repeatability)** 이 있다고 말한다. 이 개념을 시각화에 적용해보자.  

데이터를 차트에 표시하기 위해 변환한 방식이 정확하게 명시되어 있으면 이 차트는 재현 가능성이 있다. 예를 들어 어떤 사람이 만든 차트를 다른 사람이 보고 비슷하게 만들 수 있으면 재현 가능성이 있다. 반면 원본 데이터부터 화소 하나까지 다른 구석이 없는 모양새를 재현할 수 있는 시각화 도표는 반복 가능성이 있다고 한다.  

대다수 상호작용형 시각화 소프트웨어는 데이터를 변환하거나 조작하는 기능은 있어도, 최종 결과만 저장해주기에 재현 가능성이나 반복 가능성을 구현하는 것은 힘들다. 차트를 다시 만들어야 하는데, 어떻게 만들었는지 기억이 안나는 상황이 올 수도 있다. 이런 위기를 대비해 스크립트를 짜서 프로그래밍 방식으로 도표를 만드는 것이 더 좋을 수도 있다.

데이터 시각화 작업 과정은 크게 두 단계로 이루어진다.  

- **데이터 탐색(Data Exploration)**  
새로운 데이터셋을 얻었을 때, 여러 각도에서 살펴보고 다양한 시각화 방식을 시도하면서 핵심 속성을 파악해야 한다. 이 단계에서는 속도와 효율성이 관건이다. 여러 도표 유형, 다양한 데이터 변환 방식, 다양한 분류 방식을 시도해야 한다. 여러 가지 접근을 빠르게 해볼수록 더 많이 탐색적으로 분석할 수 있고, 그래야 무시될 뻔한 중요한 속성을 발견할 확률이 높아진다.  

- **데이터 표현(Data Presentation)**  
데이터를 이해하고 어떤 측면을 보여줄지 결정한 다음 돌입하는 단계로, 완성도 높으며 즉시 공개 가능한 도표를 만들어야 한다.  
  
  
탐색적 분석 단계에서는 도표를 꾸미지 않아도 된다. 데이터 표시 방식을 손쉽게 바꿀 수 있게 제작하는게 더 중요하다. 시각화 프로그램을 선택할 때는 데이터 탐색을 얼마나 빨리 할 수 있는지도 따져보는 것이 좋다.  

원하는 데이터 시각화 방식과 변환 방식, 사용할 차트 유형이 확정되면 대외적으로 공개하기 위해 완성도 높은 도표를 만들어야 한다. 여기서는 몇 가지 선택지가 있다.  
- 탐색 단계에서 사용한 소프트웨어로 최종 도표를 완성하기  
- 최종 결과물을 세련되게 만들 수 있는 소프트웨어로 바꾸기  
- 일단 시각화 소프트웨어로 시안을 만들고 포토샵이나 일러스트레이터 같은 그래픽 프로그램에서 이미지를 손수 다듬기  
- 종이와 펜, 포토샵 등을 총동원해서 도표를 다 그리기  

모든 선택지가 상황에 따라 타당하다.

좋은 소프트웨어는 도표 내용과 디자인을 따로 정할 수 있다. 내용(데이터, 데이터 변환 방식, 배치방식, 눈금, 축 범위, 차트 유형), 디자인(배경색, 전경색, 글꼴 사양, 기호 모양과 크기, 배경 격자 유무, 범례 배치, 축 눈금, 축 제목, 차트 제목 등)을 말한다.  

빠른 탐색적 분석을 통해 어떤 내용을 넣을지 구상하고, 내용을 정했으면 디자인을 수정해가면서 최적의 결과물을 만든다. 

요약하자면, 시각화 소프트웨어를 고를 때는 데이터가 바뀔 때 도표를 재현하거나 반복하기 쉬운지, 동일 데이터에 여러 시각화 방식을 빠르게 적용할 수 있는지, 도표 내용과 디자인을 별개로 수정할 수 있는지 등을 따져보자. interactive한 소프트웨어를 사용한다면, 나중에 다시 재현할 수 있도록 제작 과정을 꼼꼼하게 기록해두자.

복합 도표를 다룰 때는 구성 요소 간의 일관된 시각적 언어를 유지해야 한다. 이는 개별 도표 내에서도 마찬가지로, 하나의 이야기를 담고 있어야만 한다. 그렇지 않으면 서로 다른 결과가 머릿속에서 뒤섞여 이야기를 구별하지 못할 수 있다.

도표가 많으면 반복된 형식을 사용하되, 도표 개수도 고려해야한다. 짧은 블로그나 트위터 게시물에는 하나만 넣는게 좋고, 과학 논문에는 3~6개가 권장된다. 6개가 넘으면 일부는 부록이나 보충 자료에 넣어야할 수도 있다. 그리고 수집한 증거를 전부 기록하는건 좋지만, 비슷비슷한 도표를 여러 개 보여주면서 독자를 지치게 만들면 안된다.  

반대로 여러 가지 이야기를 한다거나, 하위 줄거리가 있는 중요한 이야기의 경우 도표를 더 많이 넣는게 적절하다.
